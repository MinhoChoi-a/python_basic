```python
import requests
from bs4 import BeautifulSoup
```

```python
res= requests.get('https://www.amazon.ca/Best-Sellers-generic/zgbs/?ref_=nav_cs_bestsellers')
```

```python
print(res.content)
```

```python
soup = BeautifulSoup(res.content, 'html.parser')
```

```python
textByClassName = soup.find_all('div', class_='zg_homeWidget')
textByClassName2 = soup.find('div', attrs={'id': 'zg_left_col1'})
```

```python
textByTag = soup.find_all('h3')
```

```python
print(textByTag)
```

```python
textByTag.get_text()
```

```python
content = textByClassName2.find_all('div', class_='zg_homeWidget')

for c in content:
    title = c.find('h3').get_text()
    
    print(title)
    
    text = c.find_all('div', attrs={'class': 'p13n-sc-truncate-desktop-type2'})
       
    for t in text:
        print(t.get_text())
```
